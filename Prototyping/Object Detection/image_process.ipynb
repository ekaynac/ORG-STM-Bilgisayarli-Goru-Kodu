```python
import os
from PIL import Image
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
PATH = r"D:\GitHub\Meturone\Meturone\UORG\Object Detection\portable_generator"
```

```python
dim = (512,512)
```

```python
resized = []
exception_paths = []
```

```python
for i in os.listdir(PATH):
    img_path = os.path.join(PATH,i)
    img = cv.imread(img_path, cv.IMREAD_UNCHANGED)
    resized_img = cv.resize(img, dim, interpolation = cv.INTER_AREA)
    try:
        cutted_img = resized_img[0:475,:,:]
    except:
        exception_paths.append(img_path)
    cutted_resized_img = cv.resize(cutted_img, dim, interpolation = cv.INTER_AREA)
    resized.append(cutted_resized_img)

resized = np.array(resized)
```

```python
plt.imshow(resized[310])
```

```python
for i in range(len(resized)):
    img = Image.fromarray(resized[i])
    name = "portable_generator" + str(i) + ".png"
    img.save(name)
```
