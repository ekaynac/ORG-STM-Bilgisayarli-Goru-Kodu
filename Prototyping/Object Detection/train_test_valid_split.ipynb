```python
import os
import shutil
```

```python
PATH = r"D:\GitHub\Meturone\Meturone\UORG\Object Detection\dataset\train"

PARENT_PATH = r"D:\GitHub\Meturone\Meturone\UORG\Object Detection\dataset"
```

```python
test_path = PARENT_PATH+r"\test"
train_path = PARENT_PATH+r"\train"
```

```python
size = int(len(os.listdir(PATH)) * 0.9)

for i in range(0,size,2):
    os.makedirs(test_path,exist_ok=True)
    print(size)
    png = os.listdir(PATH)[i]
    xml = os.listdir(PATH)[i+1]
    png_path_tar = os.path.join(test_path, png)
    xml_path_tar = os.path.join(test_path, xml)
    png_path_org = os.path.join(train_path, png)
    xml_path_org = os.path.join(train_path, xml)
    shutil.move(png_path_org, png_path_tar)
    shutil.move(xml_path_org, xml_path_tar)
```
