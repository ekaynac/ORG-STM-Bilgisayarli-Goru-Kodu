```python
import os
import shutil
```

```python
PATH = r"D:\GitHub\Meturone\Meturone\UORG\Object Detection\dataset"
```

```python
dataset_names = []
for i in os.listdir(PATH):
    img_path = os.path.join(PATH,i)
    if i[-3:] == "xml":
        dataset_names.append(i[:-4])
    
```

```python
dirr = "train"
```

```python
for name in dataset_names:
    png_path_or = os.path.join(PATH,name)+".png"
    xml_path_or = os.path.join(PATH,name)+".xml"

    train_tarr = PATH + r"\train"
    os.makedirs(train_tarr,exist_ok=True)
    png_path_tar = os.path.join(train_tarr,name)+".png"

    
    xml_path_tar = os.path.join(train_tarr,name)+".xml"

    
    
    
    shutil.copyfile(png_path_or, png_path_tar)

    shutil.copyfile(xml_path_or, xml_path_tar)
```
